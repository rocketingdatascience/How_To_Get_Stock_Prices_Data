```python
import yfinance as yf
```

```python
tesla = yf.Ticker('TSLA')
```

```python
tesla.info
```

```python
print('Price to Book:', tesla.info['priceToBook'])
print('Peg Ratio:', tesla.info['pegRatio'])
print('Book Value:', tesla.info['bookValue'])
print('Short Ratio:', tesla.info['shortRatio'])
```

```python
tesladf = tesla.history(period='1y')
```

```python
tesladf.tail(10)
```

```python
tesladf.describe()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.figure(figsize=(18,10))
plt.plot(tesladf.index, tesladf.Close, c='navy')
plt.legend(['Tesla'], loc='upper left')
plt.title('Tesla Stock Price')
plt.show()
```
